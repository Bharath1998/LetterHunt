```python
import pandas
import seaborn as sns
```

```python
df = pandas.read_csv("./gameData2.csv")
df
```

```python
box_plot_final_health = sns.boxplot(data=df, x='level', y='final_health')
box_plot_final_health.figure.savefig("box_plot_final_health.png")
box_plot_final_health
```

```python
box_plot_enemies_killed = sns.boxplot(data=df, x='level', y='enemies_killed')
box_plot_enemies_killed.figure.savefig("box_plot_enemies_killed.png")
box_plot_enemies_killed
```

```python
box_plot_time_remaining = sns.boxplot(data=df, x='level', y='time_remaining')
box_plot_time_remaining.figure.savefig("box_plot_time_remaining.png")
box_plot_time_remaining
```

```python
sns.scatterplot(data=df, x="final_health", y="enemies_killed")
```
